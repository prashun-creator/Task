#Importing libraries and dataset

```python
import pandas as pd
```

```python
keyword=pd.read_csv('/content/Keywords.csv',header=None)
keyword.head()
```

```python
keyword.columns=['KEYWORD']
```

```python
Messages=pd.read_csv('/content/Messages.csv',header=None)
Messages.head()
```

```python
Messages.columns=['MESSAGES']
```

```python
Messages.count()
```

```python
from collections import Counter
```

```python
?Counter
```

```python
unique=Counter(" ".join(Messages['MESSAGES']).split(" ")).items()
unique
```

```python
messages_count=dict(unique)
```

```python
fruits={'apple':2,'banana':6}
```

```python
messages_count
```

```python
keyword_count=[]
for items in keyword['KEYWORD']:
  for key in messages_count.keys():
    if items==key:
      keyword_count.append([items,messages_count[key]])

  

  
```

```python
keyword_count=dict(keyword_count)
keyword_count
```

```python
sorted_dict = dict(sorted(keyword_count.items(),key=lambda item: item[1],reverse=True))
sorted_dict
```

```python
import matplotlib.pyplot as plt
x=list(sorted_dict.keys())
y=list(sorted_dict.values())
plt.barh(x,y)
```
